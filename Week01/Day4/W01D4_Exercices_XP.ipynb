```python
# 1 : D√©finition des classes de base

# Classe Pets qui g√®re plusieurs animaux
class Pets():
    def __init__(self, animals):
        # animals : liste d'objets animaux
        self.animals = animals

    def walk(self):
        # Pour chaque animal de la liste, on appelle sa m√©thode walk()
        for animal in self.animals:
            print(animal.walk())

# Classe de base Cat
class Cat():
    is_lazy = True

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def walk(self):
        # M√©thode commune √† tous les chats
        return f'{self.name} se prom√®ne simplement.'
```

# R√©solution

```python
# 1 : D√©finition de la classe Dog

class Dog:
    def __init__(self, name, age, weight):
        # Initialisation des attributs de base
        self.name = name
        self.age = age
        self.weight = weight

    def bark(self):
        # M√©thode pour aboyer
        return f"{self.name} aboie !"

    def run_speed(self):
        # Calcul de la vitesse de course
        return self.weight / self.age * 10

    def fight(self, other_dog):
        # Calcul de la puissance de chaque chien
        my_power = self.run_speed() * self.weight
        other_power = other_dog.run_speed() * other_dog.weight

        # Comparaison des puissances
        if my_power > other_power:
            return f"{self.name} gagne le combat contre {other_dog.name} !"
        elif my_power < other_power:
            return f"{other_dog.name} gagne le combat contre {self.name} !"
        else:
            return "Match nul !"
```

```python
# 2 : Cr√©ation des instances de chien

# √âtape 2 : Cr√©ation de 3 chiens
dog1 = Dog("Rex", 4, 20)
dog2 = Dog("Max", 6, 25)
dog3 = Dog("Fido", 3, 10)

# V√©rification rapide
for dog in [dog1, dog2, dog3]:
    print(f"{dog.name}, {dog.age} ans, {dog.weight} kg")
```

```python
# 3 : Test des m√©thodes

# √âtape 3 : Test des m√©thodes bark(), run_speed() et fight()

# Test des aboiements
print(dog1.bark())
print(dog2.bark())
print(dog3.bark())

# Test de la vitesse
print(f"{dog1.name} court √† {dog1.run_speed():.2f} km/h")
print(f"{dog2.name} court √† {dog2.run_speed():.2f} km/h")
print(f"{dog3.name} court √† {dog3.run_speed():.2f} km/h")

# Test des combats
print(dog1.fight(dog2))
print(dog2.fight(dog3))
print(dog1.fight(dog3))
```

# R√©solution

```python
# 1 : Import des modules et d√©finition (ou import) de Dog

import random

# Pour ce notebook, on red√©finit Dog ici (normalement √† importer d‚Äôun fichier s√©par√©)
class Dog:
    def __init__(self, name, age, weight):
        self.name = name
        self.age = age
        self.weight = weight

    def bark(self):
        return f"{self.name} aboie !"

    def run_speed(self):
        return self.weight / self.age * 10

    def fight(self, other_dog):
        my_power = self.run_speed() * self.weight
        other_power = other_dog.run_speed() * other_dog.weight
        if my_power > other_power:
            return f"{self.name} gagne contre {other_dog.name} !"
        elif my_power < other_power:
            return f"{other_dog.name} gagne contre {self.name} !"
        else:
            return "Match nul !"
```

```python
# 2 : D√©finition de la classe PetDog

class PetDog(Dog):
    def __init__(self, name, age, weight):
        # Appelle le constructeur de Dog
        super().__init__(name, age, weight)
        # Attribut qui indique si le chien est entra√Æn√©
        self.trained = False

    def train(self):
        # Aboie puis devient entra√Æn√©
        print(self.bark())
        self.trained = True

    def play(self, *args):
        # args : noms des autres chiens (ou objets)
        names = [self.name] + [dog if isinstance(dog, str) else dog.name for dog in args]
        print(f"{', '.join(names)} jouent ensemble !")

    def do_a_trick(self):
        # Si entra√Æn√©, affiche une astuce al√©atoire
        if self.trained:
            tricks = [
                "fait un tonneau",
                "se met debout sur ses pattes arri√®re",
                "te serre la main",
                "fait le mort"
            ]
            print(f"{self.name} {random.choice(tricks)}")
        else:
            print(f"{self.name} n'est pas encore entra√Æn√©.")
```

```python
# 3 : Test des m√©thodes de PetDog

# Cr√©ation d'instances
dog1 = PetDog("Rex", 3, 18)
dog2 = PetDog("Max", 5, 22)
dog3 = PetDog("Buddy", 4, 20)

# Test train()
dog1.train()

# Test play() avec plusieurs chiens (on peut passer les objets ou juste les noms)
dog1.play(dog2, dog3)
# ou bien dog1.play("Max", "Buddy")

# Test do_a_trick()
dog1.do_a_trick()
dog2.do_a_trick()  # pas entra√Æn√©, donc ne fait rien
```

# R√©solution

```python
# 1 : D√©finition de la classe Person

class Person:
    def __init__(self, first_name, age):
        self.first_name = first_name
        self.age = age
        self.last_name = ""  # Initialis√© comme cha√Æne vide

    def is_18(self):
        # Retourne True si la personne a 18 ans ou plus, sinon False
        return self.age >= 18
```

```python
# 2 : D√©finition de la classe Family

class Family:
    def __init__(self, last_name):
        self.last_name = last_name
        self.members = []  # Liste vide pour stocker les Person

    def born(self, first_name, age):
        # Cr√©e un nouvel objet Person, d√©finit son nom de famille et l'ajoute √† la liste
        person = Person(first_name, age)
        person.last_name = self.last_name
        self.members.append(person)

    def check_majority(self, first_name):
        # Cherche la personne avec ce pr√©nom et v√©rifie la majorit√©
        found = False
        for person in self.members:
            if person.first_name == first_name:
                found = True
                if person.is_18():
                    print("You are over 18, your parents Jane and John accept that you will go out with your friends")
                else:
                    print("Sorry, you are not allowed to go out with your friends.")
        if not found:
            print("Person not found in the family.")

    def family_presentation(self):
        # Affiche le nom de famille et la liste des membres (pr√©nom et √¢ge)
        print(f"Nom de famille : {self.last_name}")
        print("Membres :")
        for person in self.members:
            print(f"- {person.first_name}, {person.age} ans")
```

```python
# 3 : Tests des classes Family et Person

# Cr√©ation d'une famille
smith_family = Family("Smith")

# Ajout de membres
smith_family.born("Alice", 17)
smith_family.born("Bob", 20)
smith_family.born("Charlie", 12)

# V√©rification de la majorit√© pour chaque membre
smith_family.check_majority("Alice")
smith_family.check_majority("Bob")
smith_family.check_majority("Charlie")
smith_family.check_majority("Dave")  # test personne absente

# Affichage de la pr√©sentation de la famille
smith_family.family_presentation()
```
