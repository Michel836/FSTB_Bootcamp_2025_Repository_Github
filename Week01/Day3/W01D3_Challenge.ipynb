# 1 : D√©finition de la classe Farm

```python
# D√©finition de la classe Farm (Ferme)
class Farm:
    def __init__(self, farm_name):
        # Le nom de la ferme (ex : "McDonald")
        self.name = farm_name

        # Dictionnaire vide pour stocker les animaux
        # Les cl√©s sont les types d'animaux (ex : 'cow')
        # Les valeurs sont le nombre de chaque animal (ex : 5)
        self.animals = {}

    # M√©thode pour ajouter des animaux √† la ferme
    # animal_type : le type d'animal (ex : 'cow')
    # count : combien en ajouter (par d√©faut 1)
    def add_animal(self, animal_type, count=1):
        # Si l'animal existe d√©j√†, on augmente la quantit√©
        if animal_type in self.animals:
            self.animals[animal_type] += count
        # Sinon, on l'ajoute avec la quantit√© donn√©e
        else:
            self.animals[animal_type] = count

    # M√©thode pour retourner une cha√Æne qui pr√©sente la ferme et ses animaux
    def get_info(self):
        # Commence par le nom de la ferme
        result = f"{self.name}'s farm\n\n"
        # Pour chaque animal, on ajoute une ligne avec le nom et la quantit√©
        for animal, qty in self.animals.items():
            # f"{animal:<10}" aligne le nom √† gauche sur 10 caract√®res
            result += f"{animal:<10} : {qty}\n"
        # Ajoute la phrase de la chanson
        result += "\n    E-I-E-I-0!"
        # Retourne tout le texte pr√™t √† afficher
        return result

```

# 2 : M√©thodes bonus, avec explications

```python
# M√©thode pour r√©cup√©rer les types d'animaux tri√©s (par ordre alphab√©tique)
def get_animal_types(self):
    # On r√©cup√®re les cl√©s du dictionnaire (types d'animaux) et on les trie
    return sorted(self.animals.keys())

# M√©thode pour une phrase courte pr√©sentant les animaux de la ferme
def get_short_info(self):
    # On r√©cup√®re la liste tri√©e des types d'animaux
    animals_sorted = self.get_animal_types()
    animals_names = []  # Liste pour stocker les noms √† afficher

    for animal in animals_sorted:
        # Si on a plus d'un animal, on ajoute un 's' √† la fin (pluriel en anglais)
        name = animal + 's' if self.animals[animal] > 1 else animal
        animals_names.append(name)

    # Mise en forme de la phrase selon le nombre d'animaux
    if len(animals_names) == 1:
        animal_str = animals_names[0]
    elif len(animals_names) == 2:
        # Pour deux √©l√©ments, s√©par√©s par "and"
        animal_str = ' and '.join(animals_names)
    else:
        # Pour plus de deux √©l√©ments, virgules puis "and" avant le dernier
        animal_str = ', '.join(animals_names[:-1]) + ' and ' + animals_names[-1]

    # Retourne la phrase finale
    return f"{self.name}'s farm has {animal_str}."

# On ajoute dynamiquement ces m√©thodes √† la classe Farm
Farm.get_animal_types = get_animal_types
Farm.get_short_info = get_short_info
```
