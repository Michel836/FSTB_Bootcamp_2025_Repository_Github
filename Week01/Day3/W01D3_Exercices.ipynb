```python
# DÃ©finition de la classe Zoo
class Zoo:
    def __init__(self, zoo_name):
        # Attribut : nom du zoo
        self.zoo_name = zoo_name
        # Liste pour stocker les animaux
        self.animals = []
        # Dictionnaire pour stocker les groupes (lettre : liste d'animaux)
        self.groups = {}

    # MÃ©thode pour ajouter un animal (si pas dÃ©jÃ  prÃ©sent)
    def add_animal(self, new_animal):
        if new_animal not in self.animals:
            self.animals.append(new_animal)

    # MÃ©thode pour afficher tous les animaux
    def get_animals(self):
        print("Animaux dans le zoo :", self.animals)

    # MÃ©thode pour vendre (retirer) un animal si il existe
    def sell_animal(self, animal_sold):
        if animal_sold in self.animals:
            self.animals.remove(animal_sold)

    # MÃ©thode pour trier et grouper les animaux par premiÃ¨re lettre
    def sort_animals(self):
        self.animals.sort()  # Tri alphabÃ©tique de la liste
        self.groups = {}     # RÃ©initialisation des groupes
        for animal in self.animals:
            first_letter = animal[0].upper()   # PremiÃ¨re lettre en majuscule
            if first_letter not in self.groups:
                self.groups[first_letter] = [animal]
            else:
                self.groups[first_letter].append(animal)
        return self.groups

    # MÃ©thode pour afficher les groupes
    def get_groups(self):
        for letter, animals in self.groups.items():
            print(f"{letter}: {animals}")
```
