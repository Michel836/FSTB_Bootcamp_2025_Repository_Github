```python
# Définition de la classe Cat (Chat)
class Cat:
    def __init__(self, cat_name, cat_age):
        # Attributs : nom et âge
        self.name = cat_name
        self.age = cat_age
```

```python
# Étape 1 : Création de trois objets Cat (trois chats différents)
cat1 = Cat("Minette", 5)
cat2 = Cat("Félix", 8)
cat3 = Cat("Garfield", 3)
```

```python
# Étape 2 : Fonction pour trouver le chat le plus âgé parmi les trois
def find_oldest_cat(cat1, cat2, cat3):
    # On compare les âges et on retourne le chat le plus âgé
    oldest = cat1
    if cat2.age > oldest.age:
        oldest = cat2
    if cat3.age > oldest.age:
        oldest = cat3
    return oldest
```

```python
# Étape 3 : Appel de la fonction et affichage du résultat
oldest_cat = find_oldest_cat(cat1, cat2, cat3)
print(f"Le chat le plus âgé est {oldest_cat.name}, et il a {oldest_cat.age} ans.")
```

### Instructions:

Create a Dog class with methods for barking and jumping. Instantiate dog objects, call their methods, and compare their sizes.



### Step 1: Create the Dog Class

Create a class called Dog.
In the __init__ method, take name and height as parameters and create corresponding attributes.

### Step 2: Create Dog Objects

```python
# Étape 1 : Création de la classe Dog
class Dog:
    def __init__(self, name, height):
        # Attributs : nom et taille
        self.name = name
        self.height = height

    def bark(self):
        # Méthode pour aboyer
        print(f"{self.name} aboie : woof!")

    def jump(self):
        # Méthode pour sauter (deux fois la taille du chien)
        print(f"{self.name} saute à {self.height * 2} cm de haut!")
```

```python
# Étape 2 : Création des objets chien
davids_dog = Dog("Rex", 50)
sarahs_dog = Dog("Teacup", 20)
```

```python
# Étape 3 : Affichage des détails et appel des méthodes
print(f"Le chien de David s'appelle {davids_dog.name} et mesure {davids_dog.height} cm.")
davids_dog.bark()
davids_dog.jump()

print(f"Le chien de Sarah s'appelle {sarahs_dog.name} et mesure {sarahs_dog.height} cm.")
sarahs_dog.bark()
sarahs_dog.jump()
```

```python
# Étape 4 : Comparaison des tailles
if davids_dog.height > sarahs_dog.height:
    print(f"{davids_dog.name} est plus grand que {sarahs_dog.name}.")
elif davids_dog.height < sarahs_dog.height:
    print(f"{sarahs_dog.name} est plus grand que {davids_dog.name}.")
else:
    print(f"{davids_dog.name} et {sarahs_dog.name} ont la même taille.")
```

```python
# Définition de la classe Song (Chanson)
class Song:
    # Méthode d'initialisation (prend une liste de paroles)
    def __init__(self, lyrics):
        self.lyrics = lyrics  # On sauvegarde la liste dans l'attribut lyrics

    # Méthode pour chanter la chanson (afficher chaque ligne)
    def sing_me_a_song(self):
        for line in self.lyrics:      # On parcourt chaque ligne de la chanson
            print(line)               # On affiche la ligne
```

```python
# Création d'un objet Song avec une liste de paroles
stairway = Song([
    "There’s a lady who's sure",
    "all that glitters is gold",
    "and she’s buying a stairway to heaven"
])
```

```python
# Appel de la méthode pour afficher la chanson
stairway.sing_me_a_song()
```

```python
# Définition de la classe Zoo
class Zoo:
    def __init__(self, zoo_name):
        # Attribut : nom du zoo
        self.zoo_name = zoo_name
        # Liste pour stocker les animaux
        self.animals = []
        # Dictionnaire pour stocker les groupes (lettre : liste d'animaux)
        self.groups = {}

    # Méthode pour ajouter un animal (si pas déjà présent)
    def add_animal(self, new_animal):
        if new_animal not in self.animals:
            self.animals.append(new_animal)

    # Méthode pour afficher tous les animaux
    def get_animals(self):
        print("Animaux dans le zoo :", self.animals)

    # Méthode pour vendre (retirer) un animal si il existe
    def sell_animal(self, animal_sold):
        if animal_sold in self.animals:
            self.animals.remove(animal_sold)

    # Méthode pour trier et grouper les animaux par première lettre
    def sort_animals(self):
        self.animals.sort()  # Tri alphabétique de la liste
        self.groups = {}     # Réinitialisation des groupes
        for animal in self.animals:
            first_letter = animal[0].upper()   # Première lettre en majuscule
            if first_letter not in self.groups:
                self.groups[first_letter] = [animal]
            else:
                self.groups[first_letter].append(animal)
        return self.groups

    # Méthode pour afficher les groupes
    def get_groups(self):
        for letter, animals in self.groups.items():
            print(f"{letter}: {animals}")
```

```python
# Création d'un objet Zoo
ramat_gan_safari = Zoo("Ramat Gan Safari")
```

```python
# Utilisation des méthodes pour tester
ramat_gan_safari.add_animal("Giraffe")
ramat_gan_safari.add_animal("Bear")
ramat_gan_safari.add_animal("Baboon")
ramat_gan_safari.get_animals()           # Affiche tous les animaux
ramat_gan_safari.sell_animal("Bear")     # Retire "Bear"
ramat_gan_safari.get_animals()           # Affiche après vente
ramat_gan_safari.sort_animals()          # Trie et groupe les animaux
ramat_gan_safari.get_groups()            # Affiche les groupes
```
