### Step 2: Write the game logic
1. Create a function called number_guessing_game() that contains the main logic of the game.

2. Inside the function, generate a random number between 1 and 100 using random.randint(1, 100).

random_number = random.randint(1, 100)
3. Define the maximum number of attempts the player can have, for example, 7.

4. Use a for loop to give the player a limited number of attempts (7 in this case).

5. Inside the loop, prompt the player to enter their guess using input(). Convert the input to an integer.

6. Compare the player’s guess with the random number using if-elif statements. Provide feedback based on the comparison:

If the guess is too low, print "Too low!".
If the guess is too high, print "Too high!".
If the guess is correct, print a congratulatory message and end the game with break.
7. If the player doesn’t guess correctly within the allowed attempts, print a message revealing the correct number.

```python
import random  # On importe la bibliothèque random pour générer un nombre aléatoire

# Génèrer un nombre aléatoire entre 1 et 100 inclus
nombre_secret = random.randint(1, 100)

# Fixer le nombre maximum d'essais possibles à 7
tentatives = 7

# Afficher une introduction pour l'utilisateur comme guide et regle
print("Je pense à un nombre entre 1 et 100. Tu as 7 essais.")

# Démarrer la boucle de jeu : un tour par essai
for essai in range(1, tentatives + 1):
    # Demande à l'utilisateur d'entrer un nombre pour cet essai
    essai_utilisateur = int(input(f"Essai {essai} : "))
    
    # Si l'utilisateur a trouvé le bon nombre
    if essai_utilisateur == nombre_secret:
        print("Bravo, tu as trouvé le nombre !")  # Message de victoire
        break  # On arrête la boucle, car le jeu est gagné

    # Si le nombre proposé est trop petit
    elif essai_utilisateur < nombre_secret:
        print("Trop bas 📉")
    
    # Sinon, le nombre proposé est trop grand
    else:
        print("Trop haut 📈")

# S'exécute seulement si la boucle for se termine sans break, donc sans victoire
else:
    print(f"Désolé, tu as perdu. Le nombre était : {nombre_secret}")
```

```python
import py7zr
import string
from tqdm import tqdm

# Saisie utilisateur pour plus de convivialité
archive = input("Nom de l'archive 7z (ex: mon_archive.7z) : ").strip()
debut = input("Première partie du mot de passe (avant les 2 lettres à trouver) : ").strip()
fin = input("Dernière partie du mot de passe (après les 2 lettres à trouver) : ").strip()

alphabet = string.ascii_lowercase  # ou string.ascii_letters pour minuscules et majuscules

trouve = False
total = len(alphabet) ** 2

print("Recherche du mot de passe en cours...")

for c1 in tqdm(alphabet, desc="Première lettre"):
    for c2 in alphabet:
        password = f"{debut}{c1}{c2}{fin}"
        try:
            with py7zr.SevenZipFile(archive, mode='r', password=password) as archive_file:
                archive_file.extractall(path="extrait")
                print(f"\n Mot de passe trouvé : {password}")
                print("L'archive a été extraite dans le dossier 'extrait'.")
                trouve = True
                break
        except Exception:
            continue
    if trouve:
        break

if not trouve:
    print("Aucun mot de passe valide trouvé.")

```
