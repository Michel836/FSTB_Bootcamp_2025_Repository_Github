```python
import pandas as pd

# Création du DataFrame
data = {
    'Book Title': ['The Great Gatsby', 'To Kill a Mockingbird', '1984', 'Pride and Prejudice', 'The Catcher in the Rye'],
    'Author': ['F. Scott Fitzgerald', 'Harper Lee', 'George Orwell', 'Jane Austen', 'J.D. Salinger'],
    'Genre': ['Classic', 'Classic', 'Dystopian', 'Classic', 'Classic'],
    'Price': [10.99, 8.99, 7.99, 11.99, 9.99],
    'Copies Sold': [500, 600, 800, 300, 450]
}

df = pd.DataFrame(data)
```

```python
# Regrouper par auteur et additionner les ventes
print("\n>>> Ventes totales par auteur")
print(df.groupby('Author')['Copies Sold'].sum())
```

```python
# Filtrer les livres dont le prix est supérieur à 10€
print("\n>>> Livres avec prix > 10")
print(df[df['Price'] > 10])
```

```python
# Filtrer les livres de genre "Classic"
print("\n>>> Livres de genre 'Classic'")
print(df[df['Genre'] == 'Classic'])
```

```python
# Tri par copies vendues (décroissant)
print("\n>>> Tri par copies vendues")
print(df.sort_values(by='Copies Sold', ascending=False))
```

```python
# Tri par prix (croissant)
print("\n>>> Tri par prix")
print(df.sort_values(by='Price'))
```

```python
# Résumé concis du DataFrame
print("\n>>> info()")
print(df.info())
```

```python
# Résumé statistique des colonnes numériques
print("\n>>> describe()")
print(df.describe())
```
