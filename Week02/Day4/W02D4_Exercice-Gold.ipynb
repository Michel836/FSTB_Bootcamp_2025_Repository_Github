## Exercise 2: Hierarchical Indexing
Objective: Filter and retrieve all days when the temperature exceeded 30¬∞C (86¬∞F) for cities in Canada using hierarchical indexing.
Tasks:
Create a pandas DataFrame with hierarchical indexing on ‚ÄòCountry‚Äô, ‚ÄòCity‚Äô, and ‚ÄòDate‚Äô.
Write Python code to filter the dataset for the specified condition.
Explain how hierarchical indexing simplifies multi-level data analysis.

### 1. Cr√©ation du DataFrame avec index hi√©rarchique

```python
import pandas as pd

# Donn√©es fictives
data = {
    'Temperature (¬∞F)': [85, 88, 90, 75, 89, 87, 91, 70]
}
index = pd.MultiIndex.from_tuples([
    ('Canada', 'Toronto', '2023-07-01'),
    ('Canada', 'Toronto', '2023-07-02'),
    ('Canada', 'Montreal', '2023-07-01'),
    ('Canada', 'Montreal', '2023-07-02'),
    ('USA', 'New York', '2023-07-01'),
    ('USA', 'New York', '2023-07-02'),
    ('USA', 'Chicago', '2023-07-01'),
    ('USA', 'Chicago', '2023-07-02')
], names=['Country', 'City', 'Date'])

df = pd.DataFrame(data, index=index)
```

### 2. Filtrer les jours o√π la temp√©rature d√©passe 86¬∞F pour le Canada

```python
# Filtrage : pays = Canada et temp√©rature > 86¬∞F (soit > 30¬∞C)
resultat = df.loc['Canada']
jours_chauds = resultat[resultat['Temperature (¬∞F)'] > 86]
print(jours_chauds)
```

## Exercise 3: Advanced Filtering with Hierarchical Indices
Objective: Filter rows in a hierarchical DataFrame based on specific criteria.

###  1. Cr√©ation du DataFrame avec index hi√©rarchique

```python
import pandas as pd

# Donn√©es fictives
data = {
    'Salary': [48000, 52000, 60000, 47000, 55000, 51000]
}
index = pd.MultiIndex.from_tuples([
    ('IT', 'Alice'),
    ('IT', 'Bob'),
    ('HR', 'Charlie'),
    ('HR', 'Diana'),
    ('Finance', 'Eve'),
    ('Finance', 'Frank')
], names=['Department', 'Employee'])

df = pd.DataFrame(data, index=index)
```

### 2. Filtrage : salaires > 50‚ÄØ000 dans chaque d√©partement

## Exercise 5: Box Plot of Audience and Tomato Meter Scores
Objective: Compare the distribution of Tomato Meter scores and Audience scores using a box plot.
Tasks:
Create a new DataFrame that melts the Tomato Meter and Audience scores into a long format.
Use Seaborn‚Äôs boxplot to visualize the distributions of both scores.
Customize the plot by setting an appropriate color palette and adding a title.
Display the plot.

### 1. Import des biblioth√®ques

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### 2. Cr√©ation d‚Äôun DataFrame fictif

```python
# Donn√©es fictives : scores de films MCU
df = pd.DataFrame({
    'Movie': ['Iron Man', 'Thor', 'Avengers', 'Black Panther', 'Endgame'],
    'Tomato Meter': [94, 77, 91, 96, 94],
    'Audience Score': [91, 76, 92, 79, 90]
})
```

### 3. Transformation en format long (melt)

```python
df_melted = pd.melt(df, id_vars='Movie', 
                    value_vars=['Tomato Meter', 'Audience Score'],
                    var_name='Score Type', value_name='Score')
```

### 4. Cr√©ation du box plot

```python
sns.boxplot(data=df_melted, x='Score Type', y='Score', palette='Set2')

# Titre
plt.title("Distribution des scores Tomato Meter vs Audience")

# Affichage
plt.show()
```

Ce boxplot compare visuellement la distribution des scores critiques (Tomato Meter) et audience pour les films MCU.

## Exercise 6: Pair Plot of Financial Data
Objective: Use Seaborn‚Äôs pairplot to explore relationships between financial aspects of MCU movies.
Tasks:
Select relevant columns for analysis (e.g., Production Budget, Opening Weekend, Domestic Box Office, Worldwide Box Office).
Convert financial data from string to numeric if needed.
Use Seaborn‚Äôs pairplot to visualize pairwise relationships between the selected metrics.
Set hue based on the MCU phase (e.g., mcu_phase) to identify patterns related to movie phases.
Customize the plot and display it.

### 1. Import des biblioth√®ques

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### 2. Cr√©ation d‚Äôun DataFrame fictif

```python
# Donn√©es financi√®res fictives
df = pd.DataFrame({
    'Movie': ['Iron Man', 'Thor', 'Avengers', 'Black Panther', 'Endgame'],
    'Production Budget': ['$140,000,000', '$150,000,000', '$220,000,000', '$200,000,000', '$356,000,000'],
    'Opening Weekend': ['$98,000,000', '$65,000,000', '$207,000,000', '$202,000,000', '$357,000,000'],
    'Domestic Box Office': ['$318,000,000', '$181,000,000', '$623,000,000', '$700,000,000', '$858,000,000'],
    'Worldwide Box Office': ['$585,000,000', '$449,000,000', '$1,518,000,000', '$1,347,000,000', '$2,798,000,000'],
    'MCU Phase': ['Phase 1', 'Phase 1', 'Phase 1', 'Phase 3', 'Phase 3']
})
```

### 3. Conversion des montants financiers en nombres

```python
# Fonction de nettoyage : suppression du "$" et ","
def clean_money(x):
    return pd.to_numeric(x.replace('$', '').replace(',', ''))

# Appliquer la fonction aux colonnes financi√®res
cols = ['Production Budget', 'Opening Weekend', 'Domestic Box Office', 'Worldwide Box Office']
for col in cols:
    df[col] = df[col].apply(clean_money)
```

### 4. Cr√©ation du pair plot avec coloration par phase MCU

```python
sns.pairplot(df, vars=cols, hue='MCU Phase', palette='Set1')

# Affichage
plt.show()
```

Ce pairplot montre les relations entre budget, recettes et ouverture, tout en distinguant les phases MCU. Tr√®s utile pour d√©tecter des tendances entre co√ªts et performances.
