### Rotate an image by 90 degrees

```python
from scipy.ndimage import rotate
import numpy as np
from PIL import Image

def rotate_image_90_degrees(image):
    arr = np.array(image)  # Convertir l'image PIL en array numpy
    rotated_arr = rotate(arr, 90, reshape=True)  # Rotation de 90 degrés
    return Image.fromarray(rotated_arr.astype('uint8'))  # Retourner image PIL

# Exemple d'utilisation :
rotated_img = rotate_image_90_degrees(original_image)
rotated_img.show()  # Affiche dans une fenêtre
# ou avec matplotlib :
import matplotlib.pyplot as plt
plt.imshow(rotated_img)
plt.axis('off')
plt.show()
```

```python
from scipy.ndimage import rotate
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Demande à l'utilisateur le nombre de degrés
angle = float(input("De combien de degrés veux-tu rotationner l'image ? "))

def rotate_image(image, angle):
    arr = np.array(image)
    rotated_arr = rotate(arr, angle, reshape=True)
    return Image.fromarray(rotated_arr.astype('uint8'))

rotated_img = rotate_image(original_image, angle)
plt.imshow(rotated_img)
plt.axis('off')
plt.show()
```

### Flip an image horizontally and then vertically using mirror and flip functions from ImageOps

```python
from PIL import Image, ImageOps
import matplotlib.pyplot as plt

# Flip horizontal (miroir)
flipped_horizontal = ImageOps.mirror(original_image)

# Flip vertical
flipped_vertical = ImageOps.flip(original_image)

# Affichage des deux résultats
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(flipped_horizontal)
plt.title('Flip horizontal')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(flipped_vertical)
plt.title('Flip vertical')
plt.axis('off')

plt.show()
```

### Zoom in on an image (scale by 1.2x) using .resize(...)

```python
# Zoom sur l'image (échelle 1.2x)
w, h = original_image.size
zoomed_image = original_image.resize((int(w * 1.2), int(h * 1.2)))

import matplotlib.pyplot as plt
plt.imshow(zoomed_image)
plt.title('Zoom x1.2')
plt.axis('off')
plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt

# Valeur de zoom (exemple : 2.0 = 2x)
scale = float(input("Quel facteur de zoom (ex : 2.0) ? "))
w, h = original_image.size

# Taille du crop (fenêtre centrale à zoomer)
crop_w, crop_h = int(w/scale), int(h/scale)
left = (w - crop_w) // 2
top = (h - crop_h) // 2
right = left + crop_w
bottom = top + crop_h

# Crop sur la zone centrale
cropped = original_image.crop((left, top, right, bottom))

# Redimensionne le crop à la taille d'origine
zoomed = cropped.resize((w, h), Image.LANCZOS)

# Affiche côte à côte
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(original_image)
plt.title('Originale')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(zoomed)
plt.title(f'Zoom x{scale} (crop)')
plt
```
