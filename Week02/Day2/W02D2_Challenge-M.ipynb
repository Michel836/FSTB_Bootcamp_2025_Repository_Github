### Rotate an image by 90 degrees

```python
from PIL import Image

def rotate_image_30_degrees(image):
    # Rotation dans le sens inverse des aiguilles d'une montre
    return image.rotate(30, expand=True)
```

```python
from PIL import Image
import matplotlib.pyplot as plt

# Charger l'image
image_path = 'C:/Users/chume/Desktop/Flower Color Images Dataset/flower_images/flower_images/0020.png'
original_image = Image.open(image_path)

# Appliquer la rotation
rotated_image = rotate_image_30_degrees(original_image)

# Afficher les deux images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(original_image)
plt.title("Originale")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(rotated_image)
plt.title("RotÃ©e 30Â°")
plt.axis('off')

plt.show()
```

### Flip an image horizontally and then vertically using mirror and flip functions from ImageOps

```python
from PIL import Image, ImageOps
import matplotlib.pyplot as plt

# Charger l'image
image_path = 'C:/Users/chume/Desktop/Flower Color Images Dataset/flower_images/flower_images/0020.png'
original_image = Image.open(image_path)

# Flip horizontal (miroir gauche-droite)
horizontally_flipped = ImageOps.mirror(original_image)

# Flip vertical (haut-bas)
vertically_flipped = ImageOps.flip(original_image)

# Affichage
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(original_image)
plt.title("Originale")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(horizontally_flipped)
plt.title("Flip horizontal")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(vertically_flipped)
plt.title("Flip vertical")
plt.axis('off')

plt.show()
```

### Zoom in on an image (scale by 1.2x) using .resize(...)

```python
from PIL import Image
import matplotlib.pyplot as plt

# Charger l'image
image_path = 'C:/Users/chume/Desktop/Flower Color Images Dataset/flower_images/flower_images/0020.png'
original_image = Image.open(image_path)

# Calculer la nouvelle taille (zoom 1.2x)
width, height = original_image.size
new_size = (int(width * 1.2), int(height * 1.2))

# Redimensionner (zoom avant)
zoomed_image = original_image.resize(new_size, resample=Image.LANCZOS)

# Affichage
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(original_image)
plt.title("Originale")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(zoomed_image)
plt.title("Zoom 1.2x")
plt.axis('off')

plt.show()
```

### Avec choix de l'agrandissement

```python
from PIL import Image
import matplotlib.pyplot as plt

# Demande du facteur de zoom
scale = float(input("Entrez le facteur de zoom (ex : 1.2 pour zoom, 0.8 pour rÃ©trÃ©cir) : "))

# Charger l'image
image_path = 'C:/Users/chume/Desktop/Flower Color Images Dataset/flower_images/flower_images/0020.png'
original_image = Image.open(image_path)

# Taille originale
orig_width, orig_height = original_image.size

# Redimensionner selon le facteur
zoomed_width = int(orig_width * scale)
zoomed_height = int(orig_height * scale)
zoomed_image = original_image.resize((zoomed_width, zoomed_height), resample=Image.LANCZOS)

# Si zoom > 1 : recadrer au centre
if scale >= 1.0:
    left = (zoomed_width - orig_width) // 2
    top = (zoomed_height - orig_height) // 2
    right = left + orig_width
    bottom = top + orig_height
    final_image = zoomed_image.crop((left, top, right, bottom))
# Si zoom < 1 : ajouter bordure noire (centrer image rÃ©duite)
else:
    final_image = Image.new("RGB", (orig_width, orig_height), color=(0, 0, 0))
    paste_left = (orig_width - zoomed_width) // 2
    paste_top = (orig_height - zoomed_height) // 2
    final_image.paste(zoomed_image, (paste_left, paste_top))

# Affichage
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(original_image)
plt.title("Originale")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(final_image)
plt.title(f"Zoom x{scale}")
plt.axis('off')
plt.show()
```
