## üåü Exercise 1: Understanding Data Visualization
Task: Explain why data visualization is important in data analysis.
Task: Describe the purpose of a line graph in data visualization.

### ‚úÖ **Pourquoi la visualisation de donn√©es est importante :**
Elle permet de **comprendre rapidement** les tendances, les relations et les anomalies dans les donn√©es, ce qui facilite la prise de d√©cision.

### ‚úÖ **But d‚Äôun graphique en ligne (line graph) :**
Il sert √† **montrer l‚Äô√©volution d‚Äôune variable dans le temps**, en mettant en √©vidence les tendances et les variations continues.


### Construction de graphique

```python
import matplotlib.pyplot as plt

# Donn√©es
jours = ["Lun", "Mar", "Mer", "Jeu", "Ven", "Sam", "Dim"]
temperatures = [72, 74, 76, 80, 82, 78, 75]

# Cr√©ation du graphique
plt.plot(jours, temperatures)               # tracer la ligne
plt.xlabel("Day")                           # √©tiquette axe X
plt.ylabel("Temperature (¬∞F)")              # √©tiquette axe Y
plt.title("Temperature Variation Over a Week")  # titre
plt.grid(True)                              # grille (optionnel)

plt.show()                                  # affichage
```

## Exercise 3: Visualizing Monthly Sales with a Bar Chart
Objective: Generate a bar chart using Matplotlib to visualize monthly sales data for a retail store.
Tasks:
Create a list of sales values for each month (e.g., [5000, 5500, 6200, 7000, 7500]).
Label the x-axis as ‚ÄúMonth‚Äù and the y-axis as ‚ÄúSales Amount ($)‚Äù.
Add a title to the bar chart.
Display the plot.

```python
import matplotlib.pyplot as plt

# Donn√©es
mois = ["Jan", "F√©v", "Mar", "Avr", "Mai"]
ventes = [5000, 5500, 6200, 7000, 7500]

# Cr√©ation du graphique √† barres
plt.bar(mois, ventes)

# √âtiquettes et titre
plt.xlabel("Month")
plt.ylabel("Sales Amount ($)")
plt.title("Monthly Sales Data for Retail Store")
plt.grid(axis='y')  # Grille horizontale (optionnel)

# Affichage
plt.show()
```

##  Exercise 4: Visualizing the Distribution of CGPA
Objective: Create a histogram to visualize the distribution of students‚Äô CGPA.

```python
# Importation des biblioth√®ques n√©cessaires
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Exercise 5: Comparing Anxiety Levels Across Different Genders
Objective: Use a bar plot to compare the proportion of students experiencing anxiety across different genders.
Dataset Overview: The dataset includes columns: ‚ÄòDo you have Anxiety?‚Äô and ‚ÄòChoose your gender‚Äô.
Tasks:
Import necessary libraries.
Use Seaborn to create a bar plot comparing anxiety levels across genders from the dataset df.
Customize the plot with an appropriate color palette and add a title.
Display the plot.

```python
# Importation des biblioth√®ques n√©cessaires
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Exemple de DataFrame avec des donn√©es fictives
df = pd.DataFrame({
    'Choose your gender': ['Male', 'Female', 'Female', 'Male', 'Other', 'Female', 'Male', 'Other'],
    'Do you have Anxiety?': ['Yes', 'Yes', 'No', 'No', 'Yes', 'Yes', 'Yes', 'No']
})
```

```python
# Cr√©ation du bar plot : proportion d'anxi√©t√© par genre
sns.set(style="whitegrid")
sns.countplot(data=df, x='Choose your gender', hue='Do you have Anxiety?', palette='pastel')

# Ajout d‚Äôun titre
plt.title("Anxiety Levels by Gender")

# Affichage
plt.show()
```

## Exercise 6: Exploring the Relationship Between Age and Panic Attacks
Objective: Create a scatter plot to explore the relationship between students‚Äô age and the occurrence of panic attacks.
Dataset Overview: The dataset includes columns: ‚ÄòAge‚Äô and ‚ÄòDo you have Panic Attacks?‚Äô.
Tasks:
Import necessary libraries.
Convert panic attack responses to numeric values (e.g., Yes=1, No=0).
Use Seaborn‚Äôs scatterplot to create a scatter plot with ‚ÄòAge‚Äô on the x-axis and numeric panic attack responses on the y-axis.
Customize the plot to improve readability by adding labels, a title, and adjusting point styles.
Display the plot.

```python
# Importation des biblioth√®ques
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Exemple de DataFrame avec des donn√©es fictives
df = pd.DataFrame({
    'Age': [18, 19, 20, 21, 22, 23, 24, 25],
    'Do you have Panic Attacks?': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No']
})
```

```python
# Conversion des r√©ponses en valeurs num√©riques : Yes=1, No=0
df['Panic_Attack_Numeric'] = df['Do you have Panic Attacks?'].map({'Yes': 1, 'No': 0})
```

```python
# Cr√©ation du scatter plot
sns.scatterplot(data=df, x='Age', y='Panic_Attack_Numeric', style='Do you have Panic Attacks?', s=100)

# Personnalisation
plt.title("Relation entre l'√¢ge et les crises de panique")
plt.xlabel("√Çge")
plt.ylabel("Panic Attack (1=Yes, 0=No)")
plt.yticks([0, 1], ['No', 'Yes'])

# Affichage
plt.show()
```
