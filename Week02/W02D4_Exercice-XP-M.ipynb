##  üåü Exercise 2: Creating a Line Plot for Temperature Variation
Objective: Create a simple line plot using Matplotlib that represents temperature variations over a week.
Tasks:
Use a list of temperature values for each day of the week (e.g., [72, 74, 76, 80, 82, 78, 75]).
Label the x-axis as ‚ÄúDay‚Äù and the y-axis as ‚ÄúTemperature (¬∞F)‚Äù.
Add a title to the plot.
Display the plot.

```python
import matplotlib.pyplot as plt

# Donn√©es
jours = ["Lun", "Mar", "Mer", "Jeu", "Ven", "Sam", "Dim"]
temperatures = [72, 74, 76, 80, 82, 78, 75]

# Cr√©ation du graphique
plt.plot(jours, temperatures)               # tracer la ligne
plt.xlabel("Day")                           # √©tiquette axe X
plt.ylabel("Temperature (¬∞F)")              # √©tiquette axe Y
plt.title("Temperature Variation Over a Week")  # titre
plt.grid(True)                              # grille

plt.show()                                  # affichage
```

## Exercise 3: Visualizing Monthly Sales with a Bar Chart
Objective: Generate a bar chart using Matplotlib to visualize monthly sales data for a retail store.

### M√™me exercice avec base donn√©es Student Mental Health

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Charger le fichier
df = pd.read_csv("Student Mental health.csv")

# Convertir le CGPA en num√©rique (au cas o√π ce serait au format string)
df['CGPA'] = pd.to_numeric(df['What is your CGPA?'], errors='coerce')

# Cr√©er l'histogramme
sns.histplot(df['CGPA'], bins=10, color='skyblue', kde=True)

# Personnalisation
plt.title("Distribution des CGPA des √©tudiants")
plt.xlabel("CGPA")
plt.ylabel("Nombre d'√©tudiants")
plt.show()
```

```python
print(df['What is your CGPA?'].unique())
```

### Nettoyer colonne pour extraire les valeurs num√©riques :

```python
# Nettoyage : garder uniquement les parties num√©riques avant conversion
df['CGPA_clean'] = df['What is your CGPA?'].str.extract(r'(\d+\.\d+)')
df['CGPA_clean'] = pd.to_numeric(df['CGPA_clean'], errors='coerce')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Charger le fichier
df = pd.read_csv("Student Mental health.csv")

# Nettoyer les espaces superflus
df['CGPA_category'] = df['What is your CGPA?'].str.strip()

# Compter et afficher avec un countplot
sns.countplot(x='CGPA_category', data=df, color='skyblue')
plt.title("Distribution des cat√©gories de CGPA")
plt.xlabel("Tranche de CGPA")
plt.ylabel("Nombre d'√©tudiants")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Charger le fichier
df = pd.read_csv("Student Mental health.csv")

# Convertir les r√©ponses en num√©rique : Yes = 1, No = 0
df['Anxiety'] = df['Do you have Anxiety?'].map({'Yes': 1, 'No': 0})

# Barplot : proportion d'√©tudiants avec anxi√©t√© par genre
sns.barplot(x='Choose your gender', y='Anxiety', data=df, palette='pastel')

# Personnalisation du graphique
plt.title("Proportion d'anxi√©t√© par genre")
plt.xlabel("Genre")
plt.ylabel("Proportion de 'Oui'")
plt.ylim(0, 1)
plt.show()
```

## Exercise 6: Exploring the Relationship Between Age and Panic Attacks
Objective: Create a scatter plot to explore the relationship between students‚Äô age and the occurrence of panic attacks.
Dataset Overview: The dataset includes columns: ‚ÄòAge‚Äô and ‚ÄòDo you have Panic Attacks?‚Äô.
Tasks:
Import necessary libraries.
Convert panic attack responses to numeric values (e.g., Yes=1, No=0).
Use Seaborn‚Äôs scatterplot to create a scatter plot with ‚ÄòAge‚Äô on the x-axis and numeric panic attack responses on the y-axis.
Customize the plot to improve readability by adding labels, a title, and adjusting point styles.
Display the plot.

```python
# Importer biblioth√®ques
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Cr√©er DataFrame avec des donn√©es fictives
df = pd.DataFrame({
    'Age': [18, 19, 20, 21, 22, 23, 24, 25],
    'Do you have Panic Attacks?': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No']
})
```

```python
# Conversion des r√©ponses en valeurs num√©riques : Yes=1, No=0
df['Panic_Attack_Numeric'] = df['Do you have Panic Attacks?'].map({'Yes': 1, 'No': 0})
```

```python
# Cr√©er scatter plot
sns.scatterplot(data=df, x='Age', y='Panic_Attack_Numeric', style='Do you have Panic Attacks?', s=100)

# Personnalisation
plt.title("Relation entre l'√¢ge et les crises de panique")
plt.xlabel("√Çge")
plt.ylabel("Panic Attack (1=Yes, 0=No)")
plt.yticks([0, 1], ['No', 'Yes'])

# Afficher
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Exemple de DataFrame
df['Panic'] = df['Do you have Panic Attacks?'].map({'Yes': 1, 'No': 0})

# Option 1 : barplot (moyenne de panic attacks par √¢ge ou tranche)
df['AgeGroup'] = pd.cut(df['Age'], bins=[10, 15, 20, 25, 30], labels=['10-15', '15-20', '20-25', '25-30'])
sns.barplot(x='AgeGroup', y='Panic', data=df)
plt.title("Proportion de crises de panique par tranche d'√¢ge")
plt.ylabel("Proportion de 'Oui'")
plt.xlabel("Tranche d'√¢ge")
plt.show()
```
