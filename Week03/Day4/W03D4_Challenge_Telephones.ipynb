1. Manipulation avanc√©e des donn√©es (NumPy, Pandas)
Charger et v√©rifier les donn√©es :

```python
import pandas as pd
import numpy as np

df = pd.read_csv('mobile_prices.csv')
print(df.shape)
print(df.head())
print(df.info())
```

V√©rifier la pr√©sence de valeurs manquantes :

```python
print(df.isnull().sum())
```

2. Analyse statistique avanc√©e (SciPy)
a) Corr√©lations entre variables techniques et le prix (price_range) :

```python
correlations = df.corr()['price_range'].sort_values(ascending=False)
print(correlations)
```

b) Moyennes de variables techniques par gamme de prix :

```python
group_means = df.groupby('price_range').mean()
print(group_means)
```

c) Tests statistiques (ANOVA) : influence de la RAM sur la classe de prix

```python
from scipy.stats import f_oneway

ram_0 = df[df['price_range'] == 0]['ram']
ram_1 = df[df['price_range'] == 1]['ram']
ram_2 = df[df['price_range'] == 2]['ram']
ram_3 = df[df['price_range'] == 3]['ram']

anova_result = f_oneway(ram_0, ram_1, ram_2, ram_3)
print(anova_result)
```

‚Üí Conclusion : Si la p-value est tr√®s faible, la RAM influence fortement le prix.

d) Comparaison binaire : effet de la 4G sur le prix

```python
from scipy.stats import ttest_ind

four_g_1 = df[df['four_g'] == 1]['price_range']
four_g_0 = df[df['four_g'] == 0]['price_range']
t_test_result = ttest_ind(four_g_1, four_g_0)
print(t_test_result)
```

3. Visualisation avanc√©e des donn√©es (Matplotlib)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution des prix
plt.figure(figsize=(6,4))
df['price_range'].hist(bins=4)
plt.title("R√©partition des classes de prix")
plt.xlabel("Classe de prix")
plt.ylabel("Fr√©quence")
plt.show()

# RAM vs classe de prix
plt.figure(figsize=(8,5))
sns.boxplot(x='price_range', y='ram', data=df)
plt.title("RAM selon la classe de prix")
plt.xlabel("Classe de prix")
plt.ylabel("RAM (Mo)")
plt.show()

# Batterie vs classe de prix
plt.figure(figsize=(8,5))
sns.boxplot(x='price_range', y='battery_power', data=df)
plt.title("Puissance batterie selon la classe de prix")
plt.xlabel("Classe de prix")
plt.ylabel("Batterie (mAh)")
plt.show()

# Heatmap des corr√©lations
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=False, cmap='coolwarm')
plt.title("Matrice des corr√©lations")
plt.show()
```

### Completion par rapport aux questions et details demandees et en utilisant Numpy/Scipy

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les donn√©es
df = pd.read_csv('mobile_prices.csv')

# Statistiques descriptives
mean_vals = df.mean(numeric_only=True)
median_vals = df.median(numeric_only=True)
mode_vals = df.mode(numeric_only=True).iloc[0]
range_vals = df.max(numeric_only=True) - df.min(numeric_only=True)
var_vals = df.var(numeric_only=True)
std_vals = df.std(numeric_only=True)
skew_vals = df.skew(numeric_only=True)
kurt_vals = df.kurtosis(numeric_only=True)

# Tableau r√©capitulatif
summary = pd.DataFrame({
    'Mean': mean_vals,
    'Median': median_vals,
    'Mode': mode_vals,
    'Range': range_vals,
    'Variance': var_vals,
    'Std Dev': std_vals,
    'Skewness': skew_vals,
    'Kurtosis': kurt_vals
})
display(summary)

# Corr√©lation avec price_range
correlations = df.corr(numeric_only=True)['price_range'].sort_values(ascending=False)
print("Corr√©lations avec 'price_range':\n", correlations)

# Test de corr√©lation significatif (exemple avec RAM)
corr_coef, p_val = stats.pearsonr(df['ram'], df['price_range'])
print(f"Corr√©lation ram/price_range: coef = {corr_coef:.3f}, p-value = {p_val:.3g}")

# Visualisation
plt.figure()
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Matrice de corr√©lation")
plt.show()

# Histogramme des prix
plt.figure()
df['price_range'].hist()
plt.title("R√©partition des cat√©gories de prix")
plt.xlabel("Classe de prix")
plt.ylabel("Nombre de t√©l√©phones")
plt.show()

# Boxplot RAM vs Price Range
plt.figure()
sns.boxplot(x='price_range', y='ram', data=df)
plt.title("Distribution de la RAM selon la classe de prix")
plt.show()

# Scatter RAM vs Battery
plt.figure()
sns.scatterplot(x='battery_power', y='ram', hue='price_range', data=df)
plt.title("RAM vs Batterie par classe de prix")
plt.show()
```

```python
# Calcul de la variance pour chaque variable num√©rique
variance_vals = df.var(numeric_only=True)
print("Variance de chaque variable num√©rique :\n")
print(variance_vals)
```
