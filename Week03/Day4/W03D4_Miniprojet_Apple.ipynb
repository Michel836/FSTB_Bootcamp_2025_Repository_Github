√âtape 4 : Analyse statistique avanc√©e
### Calculer les rendements journaliers :

````python
apple_data['Return'] = apple_data['Close'].pct_change()
````

### Calculer moyenne, √©cart-type, etc. :

````python
mean_return = apple_data['Return'].mean()
std_return = apple_data['Return'].std()
print("Moyenne des rendements :", mean_return)
print("√âcart-type des rendements :", std_return)
````

**Moyenne des rendements** : 0.00107
**√âcart-type des rendements** : 0.02779

### Interpr√©tation¬†:

* La moyenne est **l√©g√®rement positive** (rendement moyen journalier ‚âà 0,1‚ÄØ%).
* L‚Äô√©cart-type montre la **volatilit√©** (‚âà 2,78‚ÄØ% par jour).
* Cela sugg√®re un rendement mod√©r√© avec une volatilit√© notable, typique d‚Äôune action comme Apple.


### Test de normalit√© (ex : Shapiro-Wilk) :

````python
from scipy.stats import shapiro

result = shapiro(apple_data['Return'].dropna())
print("Statistique du test Shapiro-Wilk :", result.statistic)
print("p-value :", result.pvalue)
````

Le test Shapiro-Wilk montre¬†:

* Statistique : **0.917**
* p-value : **\~0.0** (tr√®s faible)
* **Avertissement** : le test Shapiro-Wilk n‚Äôest pas fiable pour des √©chantillons > 5000 (ici, N=11106).

**Interpr√©tation¬†:**

* p-value << 0.05 ‚áí **donn√©es non normales**.
* **Mais**¬†: r√©sultat √† interpr√©ter avec prudence √† cause de la taille de l‚Äô√©chantillon.

**R√©sum√©¬†:**
Les donn√©es ne suivent pas une loi normale selon Shapiro-Wilk, mais utilisez aussi d‚Äôautres tests/visualisations (QQ-plot, histogramme) pour confirmer.

````python
import matplotlib.pyplot as plt

# Calcul de la moyenne mobile 200 jours
apple_data['MA_200'] = apple_data['Close'].rolling(window=200).mean()

# Identifier les croisements
cross_up = (apple_data['Close'] > apple_data['MA_200']) & (apple_data['Close'].shift(1) <= apple_data['MA_200'].shift(1))
cross_down = (apple_data['Close'] < apple_data['MA_200']) & (apple_data['Close'].shift(1) >= apple_data['MA_200'].shift(1))

# Trac√©
plt.figure(figsize=(14, 6))
plt.plot(apple_data['Date'], apple_data['Close'], label='Prix de cl√¥ture', alpha=0.6)
plt.plot(apple_data['Date'], apple_data['MA_200'], label='Moyenne mobile 200j', color='green', linestyle='--')

# Marquer les croisements
plt.scatter(apple_data['Date'][cross_up], apple_data['Close'][cross_up], color='blue', marker='^', label='Croisement haussier', zorder=5)
plt.scatter(apple_data['Date'][cross_down], apple_data['Close'][cross_down], color='red', marker='v', label='Croisement baissier', zorder=5)

plt.title('Croisements entre prix de cl√¥ture et moyenne mobile 200 jours')
plt.xlabel('Date')
plt.ylabel('Prix ($)')
plt.legend()
plt.grid(True)
plt.show()
````
