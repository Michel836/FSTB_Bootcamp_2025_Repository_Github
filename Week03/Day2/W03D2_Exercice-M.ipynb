## üåü Exercise 1 : Matrix Operations
Instructions
In this exercise, you‚Äôll work with a 3x3 matrix. Here‚Äôs a brief explanation of the concepts:

Determinant: The determinant is a value that can be computed from the elements of a square matrix. It provides important information about the matrix, such as whether it has an inverse, and is used in various areas like linear algebra and calculus. To understand more about it you can watch this video.
Inverse of a Matrix: The inverse of a matrix is a matrix that, when multiplied with the original matrix, results in an identity matrix. Not all matrices have inverses. The inverse is crucial in solving systems of linear equations.
Create a 3x3 matrix and perform the following operations:

Calculate the determinant.
Find the inverse of the matrix.

### 1-Import des biblioth√®ques

```python
# Importer NumPy pour les calculs matriciels
import numpy as np
```

### 2-D√©finition de la matrice

```python
# Cr√©er une matrice 3x3
matrix = np.array([
    [4, 7, 2],
    [3, 6, 1],
    [2, 5, 3]
])

# Afficher la matrice
print("Matrice 3x3 :\n", matrix)
```

### 3-Calcul du d√©terminant

```python
# Calculer le d√©terminant de la matrice
det = np.linalg.det(matrix)

# Afficher le r√©sultat
print(f"D√©terminant de la matrice : {det:.2f}")
```

La matrice est inversible car le d√©terminant ‚â† 0.

Si le d√©terminant √©tait 0 ‚Üí la matrice serait singuli√®re (non inversible).

Ce que √ßa signifie math√©matiquement :
* L‚Äôespace engendr√© par les lignes ou colonnes est de rang complet (rang = 3 ici).

* Les vecteurs colonnes sont lin√©airement ind√©pendants.

### 4-Calcul de l'inverse

```python
# V√©rifier si la matrice est inversible
if det != 0:
    inverse = np.linalg.inv(matrix)
    print("Inverse de la matrice :\n", np.round(inverse, 3))
else:
    print("La matrice n'est pas inversible (d√©terminant nul).")
```

### 5-V√©rification (M √ó M‚Åª¬π = Identit√©)

```python
# Multiplier la matrice par son inverse pour v√©rifier si on obtient l'identit√©
if det != 0:
    identity = np.dot(matrix, inverse)
    print("Produit Matrice √ó Inverse (devrait √™tre proche de l'identit√©) :\n", np.round(identity, 3))
```

La matrice est bien inversible
Le produit donne la matrice identit√© (√† des arrondis pr√®s, -0. = 0).

* Cela confirme que le calcul de l'inverse est correct.

## üåü Exercise 2 : Statistical Analysis
Instructions
In this exercise, you‚Äôll calculate statistical measures for a dataset:

Mean: The average value of a dataset.
Median: The middle value in a dataset when it is arranged in ascending or descending order.
Standard Deviation: A measure of the amount of variation or dispersion in a set of values.
Using NumPy, generate an array of 50 random numbers and compute:

The mean and median.
The standard deviation.

### 1-Import des biblioth√®ques

```python
# Import de la biblioth√®que NumPy pour les calculs num√©riques
import numpy as np
```

### 2-G√©n√©ration de donn√©es

```python
# G√©n√©rer un tableau de 50 nombres al√©atoires entre 0 et 100
data = np.random.uniform(0, 100, size=50)

# Afficher les donn√©es g√©n√©r√©es
print("Donn√©es g√©n√©r√©es :")
print(data)
```

```python
np.size(data)
```

### 3-Calcul de la moyenne et de la m√©diane

```python
# Moyenne : somme des valeurs divis√©e par le nombre de valeurs
mean_val = np.mean(data)

# M√©diane : valeur centrale apr√®s tri des donn√©es
median_val = np.median(data)

print(f"Moyenne : {mean_val:.2f}")
print(f"M√©diane : {median_val:.2f}")
```

### 4-Calcul de l'√©cart-type

```python
# √âcart-type : mesure de la dispersion des donn√©es autour de la moyenne
std_dev = np.std(data)

print(f"√âcart-type : {std_dev:.2f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.hist(data, bins=10, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(mean_val, color='red', linestyle='dashed', linewidth=2, label=f"Moyenne: {mean_val:.2f}")
plt.axvline(median_val, color='green', linestyle='dashed', linewidth=2, label=f"M√©diane: {median_val:.2f}")
plt.title("Histogramme des donn√©es g√©n√©r√©es")
plt.xlabel("Valeurs")
plt.ylabel("Fr√©quence")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### Interpr√©tation du graphique :
* La moyenne < m√©diane : ‚Üí distribution l√©g√®rement asym√©trique √† gauche (pr√©sence de petites valeurs qui "tirent" la moyenne).

* Le pic de fr√©quence est autour de 60-70 : c‚Äôest l√† que la majorit√© des valeurs se concentrent.

## üåü Exercise 6 : Basic Hypothesis Testing
Instructions
Create a sample dataset to test the effectiveness of a new training program on employee productivity:


import numpy as np

* Productivity scores of employees before the training program
productivity_before = np.random.normal(loc=50, scale=10, size=30)

* Productivity scores of the same employees after the training program
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)

* Your task is to formulate a hypothesis regarding the training program's effectiveness 
* and test it using basic statistical functions in NumPy.


Given a dataset, formulate a simple hypothesis and test it using basic statistical functions in NumPy.

### 1-Import des biblioth√®ques

```python
import numpy as np
from scipy import stats
```

### 2-G√©n√©ration des donn√©es

```python
# Fixer un seed al√©atoir pour reproductibilit√©
np.random.seed(0)

# Scores de productivit√© avant formation (moyenne 50, √©cart-type 10)
productivity_before = np.random.normal(loc=50, scale=10, size=30)

# Scores apr√®s formation (en moyenne +5 points avec variabilit√©)
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)

# Afficher les deux √©chantillons
print("Avant formation :", np.round(productivity_before, 2))
print("Apr√®s formation :", np.round(productivity_after, 2))
```

### **Analyse rapide** de ces r√©sultats de productivit√© **avant et apr√®s** formation :

### **Tendance g√©n√©rale :**

* **Avant formation** :

  * Moyennes autour de 50‚Äì55
  * Plusieurs valeurs autour de 40 et m√™me 24.47 (basse productivit√©)

* **Apr√®s formation** :

  * Valeurs **globalement plus hautes**
  * La majorit√© des scores sont au-dessus de 55, certains atteignent **77.6**
  * √âcart r√©duit entre les employ√©s ‚Üí am√©lioration assez g√©n√©ralis√©e

---

### **Comparaisons sp√©cifiques :**

* L‚Äôemploy√© avec **24.47** est pass√© √† **26.78** ‚Üí progr√®s modeste mais r√©el
* L‚Äôemploy√© avec **40.23** est mont√© √† **45.70**
* Beaucoup d‚Äôemploy√©s autour de **65‚Äì70** apr√®s formation (versus moins nombreux avant)

---

### **Conclusion :**

* Il y a un **gain net et r√©gulier** de performance.
* Cela confirme visuellement que la formation a eu un **effet positif g√©n√©ralis√©**.
* La **dispersion diminue** l√©g√®rement et les scores **montent globalement**.


### 3-Formulation et test de l‚Äôhypoth√®se

```python
# H0 : pas de diff√©rence (formation inefficace)
# H1 : la formation am√©liore la productivit√© (moyenne apr√®s > moyenne avant)

# Test t pour donn√©es appari√©es
t_stat, p_val = stats.ttest_rel(productivity_after, productivity_before)

# Test unilat√©ral (am√©lioration ‚Üí moyenne apr√®s > avant)
p_val_one_tailed = p_val / 2 if t_stat > 0 else 1 - (p_val / 2)

# Moyennes pour information
mean_before = np.mean(productivity_before)
mean_after = np.mean(productivity_after)

print(f"Moyenne avant : {mean_before:.2f}")
print(f"Moyenne apr√®s : {mean_after:.2f}")
print(f"Statistique t : {t_stat:.2f}")
print(f"p-valeur unilat√©rale : {p_val_one_tailed:.2e}")
```

### Analyse
* La formation am√©liore significativement la productivit√© des employ√©s.
* Les r√©sultats sont √† la fois statistiquement significatifs et coh√©rents en moyenne.

### 4-Interpr√©tation

```python
if p_val_one_tailed < 0.05:
    print("R√©sultat : la formation a un effet statistiquement significatif.")
else:
    print("R√©sultat : pas de preuve suffisante que la formation a eu un effet.")
```

### 5-Visualisation

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(productivity_before, label="Avant", marker='o')
plt.plot(productivity_after, label="Apr√®s", marker='o')
plt.title("Productivit√© avant/apr√®s formation")
plt.ylabel("Score de productivit√©")
plt.xlabel("Employ√©")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### 6-Repr√©sentations alternatives

```python
import matplotlib.pyplot as plt

plt.boxplot([productivity_before, productivity_after], labels=["Avant", "Apr√®s"])
plt.title("Distribution de la productivit√© (avant vs apr√®s formation)")
plt.ylabel("Score de productivit√©")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
means = [np.mean(productivity_before), np.mean(productivity_after)]
stds = [np.std(productivity_before), np.std(productivity_after)]

plt.bar(["Avant", "Apr√®s"], means, yerr=stds, capsize=8, color=["gray", "skyblue"])
plt.title("Moyenne de productivit√© avant/apr√®s formation")
plt.ylabel("Score moyen")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

Conclusion :
* Le boxplot ou le graphique en barres avec erreur sont bien plus adapt√©s pour interpr√©ter visuellement un test d'hypoth√®se sur deux groupes.
