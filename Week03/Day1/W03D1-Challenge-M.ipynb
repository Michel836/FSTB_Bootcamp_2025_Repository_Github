```python
# Import des bibliothèques nécessaires
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 1. Génération des données synthétiques

```python

# Noms des villes et mois
villes = ["Paris", "Londres", "New York", "Tokyo", "Sydney", "Moscou", "Le Cap", "Delhi", "Rio", "Toronto"]
mois = ["Jan", "Fév", "Mar", "Avr", "Mai", "Juin", "Juil", "Août", "Sep", "Oct", "Nov", "Déc"]
```

```python
# Génération d'un tableau 10x12 de températures entre -5 et 35
temperatures = np.random.uniform(low=-5, high=35, size=(len(villes), len(mois)))
```

```python
# Conversion en DataFrame pandas
df_temp = pd.DataFrame(temperatures, index=villes, columns=mois)
```

```python
# Affichage rapide pour vérification
print(df_temp.head())
```

# 2. Analyse des données

```python
# Calcul de la température moyenne annuelle par ville
df_temp["Moyenne annuelle"] = df_temp.mean(axis=1)
```

```python
# Ville la plus chaude en moyenne annuelle
ville_max = df_temp["Moyenne annuelle"].idxmax()
temp_max = df_temp.loc[ville_max, "Moyenne annuelle"]
```

```python
# Ville la plus froide en moyenne annuelle
ville_min = df_temp["Moyenne annuelle"].idxmin()
temp_min = df_temp.loc[ville_min, "Moyenne annuelle"]

print(f"\nVille la plus chaude en moyenne annuelle : {ville_max} ({temp_max:.2f}°C)")
print(f"Ville la plus froide en moyenne annuelle : {ville_min} ({temp_min:.2f}°C)")
```

# 3. Visualisation

```python
plt.figure(figsize=(14, 7))
for ville in villes:
    plt.plot(mois, df_temp.loc[ville, mois], label=ville)

plt.title("Températures mensuelles moyennes par ville (°C)")
plt.xlabel("Mois")
plt.ylabel("Température (°C)")
plt.legend(loc="upper right")
plt.grid(True)
plt.show()
```

# 4. Rapport d'analyse des températures moyennes mensuelles pour 10 villes et visuelle du graphique :
