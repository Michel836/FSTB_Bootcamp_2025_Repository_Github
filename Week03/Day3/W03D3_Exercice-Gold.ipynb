## Exercice 1 : Distribution normale multivari√©e
* T√¢che : G√©n√©rer et visualiser une distribution normale univari√©e et multivari√©e √† l‚Äôaide de SciPy. Expliquez leurs diff√©rences.

* Astuce pour la distribution univari√©e :

* Utilisation : norm √† partir du module de SciPy stats.
Exemple de code :
from scipy.stats import norm
univariate_data = norm.rvs(size=1000)
Astuce pour la distribution multivari√©e :
Utilisation : multivariate_normal du m√™me module.
Exemple de code :
from scipy.stats import multivariate_normal
mean = [0, 0]
cov = [[1, 0], [0, 1]]  # diagonal covariance, independent variables
multivariate_data = multivariate_normal.rvs(mean=mean, cov=cov, size=1000)
Explication de la t√¢che : Comparer les distributions univari√©es et multivari√©es g√©n√©r√©es. Expliquez comment la pr√©sence de plusieurs variables et leurs corr√©lations dans le cas multivari√© entra√Ænent des diff√©rences de forme et de dispersion de la distribution.

```python
# Importation des biblioth√®ques n√©cessaires
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, multivariate_normal
```

```python
# G√©n√©ration de 1000 points d'une distribution normale univari√©e (1 variable)
univariate_data = norm.rvs(size=1000)
```

```python
# Visualisation : histogramme de la distribution univari√©e
plt.figure(figsize=(8,5))
plt.hist(univariate_data, bins=30, density=True, alpha=0.7)
plt.title("Distribution normale univari√©e")
plt.xlabel("Valeurs")
plt.ylabel("Densit√©")
plt.grid(True)
plt.show()
```

```python
# Param√®tres pour la distribution normale multivari√©e (2 variables)
mean = [0, 0]  # moyenne pour X et Y
cov = [[1, 0], [0, 1]]  # covariance diagonale : X et Y ind√©pendants
```

```python
# G√©n√©ration de 1000 points (2 dimensions)
multivariate_data = multivariate_normal.rvs(mean=mean, cov=cov, size=1000)
```

```python
# S√©paration des colonnes X et Y pour tracer
x = multivariate_data[:, 0]
y = multivariate_data[:, 1]
```

```python
# Visualisation : nuage de points de la distribution multivari√©e
plt.figure(figsize=(6,6))
plt.scatter(x, y, alpha=0.5)
plt.title("Distribution normale multivari√©e (X,Y)")
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.axis('equal')
plt.grid(True)
plt.show()
```

## Diff√©rences entre les deux distributions :

* Univari√©e : suit une courbe en cloche (loi normale) avec une seule variable. 
   Elle est repr√©sent√©e par un histogramme en 1D.

* Multivari√©e : suit une loi normale dans un espace √† plusieurs dimensions.
   Ici, deux variables X et Y ind√©pendantes : nuage de points circulaire.

   Si les variables √©taient corr√©l√©es (covariance ‚â† 0), on aurait un nuage elliptique.

* La pr√©sence de plusieurs variables dans le cas multivari√© permet de mod√©liser
la relation conjointe entre variables, ce qui n'est pas possible avec une distribution univari√©e.

## Exercice 3 : Tests d'hypoth√®ses statistiques
* T√¢che : Effectuer et interpr√©ter un test ANOVA √† l‚Äôaide de SciPy pour comparer plus de deux groupes.
Groupes √† utiliser : vous pouvez utiliser les donn√©es fournies √† partir du code suivant :
import numpy as np
import pandas as pd

*  Seed for reproducibility
np.random.seed(0)
*  Sales data across different regions
region1 = np.random.normal(20000, 3000, 30)  # Region 1 sales
region2 = np.random.normal(22000, 3500, 30)  # Region 2 sales
region3 = np.random.normal(25000, 5000, 30)  # Region 3 sales
sales_data = pd.DataFrame({
    'Region 1': region1,
    'Region 2': region2,
    'Region 3': region3
})
* Ressource suppl√©mentaire : Voici un lien vers un guide sur le test ANOVA . Cela vous aidera √† comprendre le test, car il n'a pas encore √©t√© abord√© dans vos cours.

```python
# Importation des biblioth√®ques n√©cessaires
import numpy as np
import pandas as pd
from scipy.stats import f_oneway
import matplotlib.pyplot as plt
```

```python
# G√©n√©ration de donn√©es simul√©es pour 3 r√©gions
np.random.seed(0)
region1 = np.random.normal(20000, 3000, 30)  # Ventes r√©gion 1
region2 = np.random.normal(22000, 3500, 30)  # Ventes r√©gion 2
region3 = np.random.normal(25000, 5000, 30)  # Ventes r√©gion 3
```

```python
# Cr√©ation d'un DataFrame pour afficher les donn√©es
sales_data = pd.DataFrame({
    'Region 1': region1,
    'Region 2': region2,
    'Region 3': region3
})
sales_data.head()
```

### 1. Visualisation simple des diff√©rences

```python
# Bo√Æte √† moustaches pour comparer visuellement les distributions des ventes
sales_data.boxplot(figsize=(8, 5))
plt.title("Comparaison des ventes par r√©gion")
plt.ylabel("Ventes")
plt.grid(True)
plt.show()
```

### 2. Test ANOVA (analyse de la variance)

```python
# Application du test ANOVA unidirectionnel (one-way)
f_stat, p_value = f_oneway(region1, region2, region3)

# R√©sultats
print("Statistique F :", f_stat)
print("Valeur p :", p_value)
```

### 3. Interpr√©tation

```python
alpha = 0.05  # seuil de signification

if p_value < alpha:
    print("‚Üí Il existe une diff√©rence significative entre au moins deux groupes de r√©gions (p < 0.05).")
else:
    print("‚Üí Aucune diff√©rence significative entre les moyennes des groupes (p ‚â• 0.05).")
```

### R√©sum√© :
* Objectif du test ANOVA : v√©rifier si au moins un groupe a une moyenne statistiquement diff√©rente des autres.

* Hypoth√®se nulle (H‚ÇÄ) : toutes les moyennes sont √©gales.

* Rejet de H‚ÇÄ si p < 0.05 ‚Üí preuve d'une diff√©rence significative.

* Sinon ‚Üí diff√©rences non significatives selon les donn√©es disponibles.

## Exercice 4 : Analyse de r√©gression lin√©aire
* T√¢che : impl√©menter une analyse de r√©gression lin√©aire simple √† l‚Äôaide de SciPy sur l‚Äôensemble de donn√©es fourni et interpr√©ter les r√©sultats.
* Code de g√©n√©ration de l'ensemble de donn√©es :
import numpy as np

* sd for reproducibility
np.random.seed(0)

* ample data

* apes √† suivre :
Pr√©paration des donn√©es :
Variable d√©pendante (Y) : ¬´ R√©sultat du test ¬ª
Variable ind√©pendante (X) : ¬´ Heures √©tudi√©es ¬ª
Ex√©cution de la r√©gression : utilisez la linregressfonction du statsmodule SciPy.
Interpr√©tation : Concentrez-vous sur la pente, l‚Äôinterception et les valeurs R au carr√© pour comprendre la relation entre les heures d‚Äô√©tude et les r√©sultats aux tests.

```python
# Importation des biblioth√®ques n√©cessaires
import numpy as np
import pandas as pd
from scipy.stats import linregress
import matplotlib.pyplot as plt
```

1. G√©n√©ration et pr√©paration des donn√©es

```python
# G√©n√©ration de donn√©es simul√©es
np.random.seed(0)

X = np.random.rand(100) * 50  # Heures √©tudi√©es
Y = 2.5 * X + np.random.randn(100) * 10  # R√©sultat du test (avec bruit)
```

```python
# Cr√©ation du DataFrame
linear_regression_data = pd.DataFrame({
    'Hours Studied': X,
    'Test Score': Y
})
linear_regression_data.head()
```

2. R√©gression lin√©aire avec SciPy

```python
# R√©gression lin√©aire : Y = aX + b
slope, intercept, r_value, p_value, std_err = linregress(X, Y)
```

```python
# Affichage des r√©sultats
print("Pente (slope) :", slope)
print("Interception :", intercept)
print("Coefficient de corr√©lation R :", r_value)
print("R¬≤ :", r_value**2)
print("Valeur p :", p_value)
print("Erreur standard :", std_err)
```

3. Visualisation des r√©sultats

```python
# Trac√© des donn√©es et de la droite de r√©gression
plt.figure(figsize=(8,5))
plt.scatter(X, Y, label="Donn√©es r√©elles", alpha=0.6)
plt.plot(X, slope * X + intercept, color='red', label="R√©gression")
plt.xlabel("Heures √©tudi√©es")
plt.ylabel("R√©sultat du test")
plt.title("R√©gression lin√©aire : Heures √©tudi√©es vs R√©sultat du test")
plt.legend()
plt.grid(True)
plt.show()
```

4. Interpr√©tation

‚Üí La pente de {slope:.2f} signifie que chaque heure d'√©tude suppl√©mentaire
   augmente en moyenne le score au test de {slope:.2f} points.

‚Üí L'interception de {intercept:.2f} repr√©sente le score estim√© si aucune heure n'√©tait √©tudi√©e.

‚Üí Le coefficient R¬≤ de {r_value**2:.2f} indique que {r_value**2:.0%} de la variance
   des r√©sultats est expliqu√©e par le nombre d'heures √©tudi√©es.

‚Üí Conclusion : Il existe une relation positive et significative entre le temps d‚Äô√©tude
   et le r√©sultat obtenu.
