# 1-Analyse statistique avanc√©e des donn√©es boursi√®res d'Apple Inc. :

Analyse des donn√©es financi√®res : r√©alisation d'une analyse approfondie des donn√©es boursi√®res d'Apple Inc. √† l'aide de NumPy et SciPy.
Tests d‚Äôhypoth√®ses : application de tests statistiques pour tirer des conclusions significatives √† partir de donn√©es financi√®res.
Visualisation des tendances : utilisation de Matplotlib pour visualiser les tendances et les mod√®les financiers.

√âtape 1 : R√©cup√©ration et chargement des donn√©es
Trouver et t√©l√©charger un fichier CSV des donn√©es boursi√®res d‚ÄôApple (exemple : Yahoo Finance).

Charger les donn√©es en Python :

```python
import pandas as pd

apple_data = pd.read_csv('apple_stock.csv') 
```

√âtape 2 : Pr√©traitement des donn√©es
V√©rifier la structure et les premi√®res lignes :

```python
print(apple_data.head())
print(apple_data.info())
```

Nettoyer les donn√©es :

G√©rer les valeurs manquantes

Convertir la colonne ‚ÄòDate‚Äô au bon format

Renommer la colonne unnamed en Date

```python
apple_data = apple_data.rename(columns={'Unnamed: 0': 'Date'})
apple_data['Date'] = pd.to_datetime(apple_data['Date'])
```

```python
apple_data = apple_data.dropna()
apple_data['Date'] = pd.to_datetime(apple_data['Date'])
```

√âtape 3 : Exploration rapide
Afficher des statistiques descriptives :

```python
print(apple_data.describe())
```

Visualiser le cours de cl√¥ture :

```python
import matplotlib.pyplot as plt

plt.plot(apple_data['Date'], apple_data['Close'])
plt.title("Cours de cl√¥ture Apple")
plt.xlabel("Date")
plt.ylabel("Prix de cl√¥ture")
plt.show()
```

√âtape 4 : Analyse statistique avanc√©e
Calculer les rendements journaliers :

```python
apple_data['Return'] = apple_data['Close'].pct_change()
```

Calculer moyenne, √©cart-type, etc. :

```python
mean_return = apple_data['Return'].mean()
std_return = apple_data['Return'].std()
print("Moyenne des rendements :", mean_return)
print("√âcart-type des rendements :", std_return)
```

**Moyenne des rendements** : 0.00107
**√âcart-type des rendements** : 0.02779

**Interpr√©tation¬†:**

* La moyenne est **l√©g√®rement positive** (rendement moyen journalier ‚âà 0,1‚ÄØ%).
* L‚Äô√©cart-type montre la **volatilit√©** (‚âà 2,78‚ÄØ% par jour).
* Cela sugg√®re un rendement mod√©r√© avec une volatilit√© notable, typique d‚Äôune action comme Apple.


Test de normalit√© (ex : Shapiro-Wilk) :

```python
from scipy.stats import shapiro

result = shapiro(apple_data['Return'].dropna())
print("Statistique du test Shapiro-Wilk :", result.statistic)
print("p-value :", result.pvalue)
```

Votre test Shapiro-Wilk montre¬†:

* Statistique : **0.917**
* p-value : **\~0.0** (tr√®s faible)
* **Avertissement** : le test Shapiro-Wilk n‚Äôest pas fiable pour des √©chantillons > 5000 (ici, N=11106).

**Interpr√©tation¬†:**

* p-value << 0.05 ‚áí **donn√©es non normales**.
* **Mais**¬†: r√©sultat √† interpr√©ter avec prudence √† cause de la taille de l‚Äô√©chantillon.

**R√©sum√©¬†:**
Vos donn√©es ne suivent pas une loi normale selon Shapiro-Wilk, mais utilisez aussi d‚Äôautres tests/visualisations (QQ-plot, histogramme) pour confirmer.


√âtape 5 : Visualisation avanc√©e
Visualiser l‚Äô√©volution des rendements :

```python
plt.plot(apple_data['Date'], apple_data['Return'])
plt.title("Rendements journaliers Apple")
plt.xlabel("Date")
plt.ylabel("Rendement")
plt.show()
```

Histogramme des rendements :

```python
apple_data['Return'].hist(bins=30)
plt.title("Distribution des rendements journaliers Apple")
plt.xlabel("Rendement")
plt.ylabel("Fr√©quence")
plt.show()
```

1. Charger et pr√©traiter les donn√©es

Nettoyer les valeurs manquantes, v√©rifier les types de colonnes.

1. Manipulation avanc√©e des donn√©es (NumPy, Pandas)
Charger et v√©rifier les donn√©es :

```python
import pandas as pd
import numpy as np

df = pd.read_csv('mobile_prices.csv')
print(df.shape)
print(df.head())
print(df.info())
```

V√©rifier la pr√©sence de valeurs manquantes :

```python
print(df.isnull().sum())
```

2. Analyse statistique avanc√©e (SciPy)
a) Corr√©lations entre variables techniques et le prix (price_range) :

```python
correlations = df.corr()['price_range'].sort_values(ascending=False)
print(correlations)
```

b) Moyennes de variables techniques par gamme de prix :

```python
group_means = df.groupby('price_range').mean()
print(group_means)
```

c) Tests statistiques (ANOVA) : influence de la RAM sur la classe de prix

```python
from scipy.stats import f_oneway

ram_0 = df[df['price_range'] == 0]['ram']
ram_1 = df[df['price_range'] == 1]['ram']
ram_2 = df[df['price_range'] == 2]['ram']
ram_3 = df[df['price_range'] == 3]['ram']

anova_result = f_oneway(ram_0, ram_1, ram_2, ram_3)
print(anova_result)
```

‚Üí Conclusion : Si la p-value est tr√®s faible, la RAM influence fortement le prix.

d) Comparaison binaire : effet de la 4G sur le prix

```python
from scipy.stats import ttest_ind

four_g_1 = df[df['four_g'] == 1]['price_range']
four_g_0 = df[df['four_g'] == 0]['price_range']
t_test_result = ttest_ind(four_g_1, four_g_0)
print(t_test_result)
```

3. Visualisation avanc√©e des donn√©es (Matplotlib)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution des prix
plt.figure(figsize=(6,4))
df['price_range'].hist(bins=4)
plt.title("R√©partition des classes de prix")
plt.xlabel("Classe de prix")
plt.ylabel("Fr√©quence")
plt.show()

# RAM vs classe de prix
plt.figure(figsize=(8,5))
sns.boxplot(x='price_range', y='ram', data=df)
plt.title("RAM selon la classe de prix")
plt.xlabel("Classe de prix")
plt.ylabel("RAM (Mo)")
plt.show()

# Batterie vs classe de prix
plt.figure(figsize=(8,5))
sns.boxplot(x='price_range', y='battery_power', data=df)
plt.title("Puissance batterie selon la classe de prix")
plt.xlabel("Classe de prix")
plt.ylabel("Batterie (mAh)")
plt.show()

# Heatmap des corr√©lations
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=False, cmap='coolwarm')
plt.title("Matrice des corr√©lations")
plt.show()
```

Le graphique de la repartition des classes de prix montre que la variable `price_range` (classe de prix) est **parfaitement √©quilibr√©e**‚ÄØ:

* Il y a **autant d‚Äôexemples** pour chaque classe (0, 1, 2, 3), soit environ 500 observations par classe.

**Interpr√©tation directe :**

* Cela facilite les analyses statistiques‚ÄØ: pas de biais de d√©s√©quilibre de classes.
* Toutes les classes de prix sont repr√©sent√©es √©quitablement dans ta base.

**Conclusion :**

* Tu peux comparer sans probl√®me les caract√©ristiques entre les classes, tes tests et visualisations seront fiables.
* C‚Äôest une tr√®s bonne base pour des analyses comparatives ou de la mod√©lisation supervis√©e.


Synth√®se et conclusion
La RAM et la batterie sont fortement corr√©l√©es avec la classe de prix‚ÄØ: plus ces valeurs sont √©lev√©es, plus le t√©l√©phone est cher.

La 4G a un impact significatif sur le prix (test t tr√®s significatif).

D‚Äôautres variables comme la r√©solution √©cran, le nombre de c≈ìurs, etc., influencent aussi le prix, mais moins fortement que la RAM.
