## üåü Exercice 2 : Statistiques descriptives
* T√¢che : √âtant donn√© un exemple d‚Äôensemble de donn√©es, calculez la moyenne, la m√©diane, la variance et l‚Äô√©cart type √† l‚Äôaide de SciPy.
* Exemple de jeu de donn√©es :
data = [12, 15, 13, 12, 18, 20, 22, 21]


```python
# Import des fonctions statistiques depuis scipy
from scipy import stats
import numpy as np  # Utilis√© ici pour la variance et l'√©cart type
```

```python
# Jeu de donn√©es fourni
data = [12, 15, 13, 12, 18, 20, 22, 21]
```

```python
# Moyenne : somme des valeurs divis√©e par le nombre de valeurs
mean = stats.tmean(data)
print("Moyenne :", mean)
```

```python
# M√©diane : valeur centrale quand les donn√©es sont tri√©es
median = stats.scoreatpercentile(data, 50)
print("M√©diane :", median)
```

```python
# Variance : mesure de la dispersion des donn√©es autour de la moyenne
variance = np.var(data, ddof=0)  # ddof=0 ‚Üí variance population
print("Variance :", variance)
```

```python
# √âcart type : racine carr√©e de la variance
std_dev = np.std(data, ddof=0)  # ddof=0 ‚Üí √©cart type population
print("√âcart type :", std_dev)
```

## üåü Exercice 3 : Comprendre les distributions
* T√¢che : G√©n√©rer une distribution normale √† l'aide de SciPy avec une moyenne de 50 et un √©cart type de 10. Tracer cette distribution.
* Exemple de code :
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

* complete your code here


```python
# Importation des biblioth√®ques n√©cessaires
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Param√®tres de la distribution normale
mu = 50         # moyenne
sigma = 10      # √©cart type
```

```python
# G√©n√©ration d'une plage de valeurs autour de la moyenne
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
```

```python
# Calcul de la densit√© de probabilit√© pour chaque valeur de x
y = norm.pdf(x, mu, sigma)
```

```python
# Trac√© de la courbe de la distribution normale
plt.plot(x, y)
plt.title("Distribution Normale (Œº=50, œÉ=10)")
plt.xlabel("Valeurs")
plt.ylabel("Densit√© de probabilit√©")
plt.grid(True)
plt.show()
```

## üåü Exercice 4 : Application du test T
* T√¢che : Effectuer un test T sur deux ensembles de donn√©es g√©n√©r√©es al√©atoirement.
* Exemple de code :
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

* complete your code here


```python
# Importation des biblioth√®ques n√©cessaires
import numpy as np
from scipy import stats
```

```python
# G√©n√©ration de deux √©chantillons al√©atoires suivant une distribution normale
# data1 : moyenne 50, √©cart type 10, 100 valeurs
# data2 : moyenne 60, √©cart type 10, 100 valeurs
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)
```

```python
# Test t de Student pour √©chantillons ind√©pendants
# H0 : les deux √©chantillons proviennent de populations avec la m√™me moyenne
t_stat, p_value = stats.ttest_ind(data1, data2)
```

```python
# Affichage des r√©sultats du test
print("Statistique t :", t_stat)
print("Valeur p :", p_value)
```


* Une p-value < 0.05 sugg√®re un rejet de H0 (diff√©rence significative entre les moyennes)

## Exercice 5 : Analyse de r√©gression lin√©aire
* Objectif : Appliquer la r√©gression lin√©aire √† un ensemble de donn√©es et interpr√©ter les r√©sultats.

* T√¢che : √âtant donn√© un ensemble de donn√©es sur les prix des logements ( house_prices) et leurs tailles correspondantes ( house_sizes), utilisez la r√©gression lin√©aire pour pr√©dire le prix d'une maison en fonction de sa taille.
* Ensemble de donn√©es :
house_sizes: [50, 70, 80, 100, 120] (en m√®tres carr√©s)
house_prices: [150 000, 200 000, 210 000, 250 000, 280 000] (en unit√©s mon√©taires)
* Questions :
Quelle est la pente et l'interception de la droite de r√©gression ?
Pr√©disez le prix d'une maison de 90 m√®tres carr√©s.
Interpr√©ter la signification de la pente dans le contexte des prix de l‚Äôimmobilier.

```python
# Importation des biblioth√®ques n√©cessaires
import numpy as np
from scipy import stats
```

```python
# Donn√©es : tailles des maisons (m¬≤) et prix correspondants
house_sizes = np.array([50, 70, 80, 100, 120])
house_prices = np.array([150000, 200000, 210000, 250000, 280000])
```

```python
# R√©gression lin√©aire : calcule la pente, l'interception, r, p-value, erreur std
slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)
```

```python
# Affichage des coefficients
print("Pente :", slope)
print("Interception :", intercept)
```

```python
# Pr√©diction du prix pour une maison de 90 m¬≤
predicted_price = slope * 90 + intercept
print("Prix pr√©dit pour une maison de 90 m¬≤ :", predicted_price)
```

### Interpr√©tation :
* La pente repr√©sente l'augmentation moyenne du prix pour chaque m√®tre carr√© suppl√©mentaire.
* Ici, chaque m¬≤ suppl√©mentaire fait augmenter le prix d'environ `slope` unit√©s mon√©taires.

## üåü Exercice 6 : Comprendre l‚ÄôANOVA
* Objectif : Tester la compr√©hension de l'ANOVA et de son application.

* T√¢che : Trois engrais diff√©rents sont appliqu√©s √† trois groupes de plantes distincts pour tester leur efficacit√©. La croissance en centim√®tres est enregistr√©e.
* Ensemble de donn√©es :
fertilizer_1: [5, 6, 7, 6, 5]
fertilizer_2: [7, 8, 7, 9, 8]
fertilizer_3: [4, 5, 4, 3, 4]
* Questions :
Effectuez un test ANOVA sur les donn√©es fournies. Quelles sont les valeurs F et P ?
Sur la base de la valeur P, les engrais ont-ils des effets significativement diff√©rents sur la croissance des plantes ?
Expliquez ce qui se passerait si la valeur p √©tait sup√©rieure √† 0,05.

```python
# Importation de la fonction ANOVA (analyse de la variance) et numpy
from scipy import stats
import numpy as np
```

```python
# Donn√©es de croissance des plantes selon 3 types d'engrais
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]
```

```python
# Application du test ANOVA √† un facteur (one-way ANOVA)
f_stat, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)
```

```python
# Affichage des r√©sultats
print("Valeur F :", f_stat)
print("Valeur p :", p_value)
```

### Interpr√©tation :
* Si p < 0.05 : on rejette l'hypoth√®se nulle ‚Üí au moins un engrais a un effet diff√©rent
* Si p > 0.05 : on ne peut pas conclure √† une diff√©rence significative entre les engrais

```python
if p_value < 0.05:
    print("R√©sultat : Les engrais ont des effets significativement diff√©rents.")
else:
    print("R√©sultat : Aucune diff√©rence significative d√©tect√©e entre les engrais.")
```

## Exercice 7 : Distributions de probabilit√© (facultatif)
* Travailler avec une distribution binomiale : calculer les probabilit√©s pour diff√©rents nombres de succ√®s.
* Exemple : Calculer la probabilit√© d‚Äôobtenir exactement 5 faces en 10 lancers de pi√®ce.

```python
# Importation de la distribution binomiale depuis scipy
from scipy.stats import binom
```

```python
# Param√®tres de la distribution binomiale
n = 10      # nombre d'essais (lancers de pi√®ce)
p = 0.5     # probabilit√© de succ√®s (obtenir face)
```

```python
# Calcul de la probabilit√© d'obtenir exactement 5 faces
k = 5
prob = binom.pmf(k, n, p)  # pmf = probabilit√© de masse (discr√®te)

print(f"Probabilit√© d‚Äôobtenir exactement {k} faces sur {n} lancers :", prob)
```

## üåü Exercice 8 : Coefficients de corr√©lation (facultatif)
* Calculez les coefficients de corr√©lation de Pearson et de Spearman entre deux variables dans un ensemble de donn√©es.
* Exemple dans le code :data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})


```python
# Importation des biblioth√®ques n√©cessaires
import pandas as pd
from scipy.stats import pearsonr, spearmanr
```

```python
# Cr√©ation de l'ensemble de donn√©es
data = pd.DataFrame({
    'age': [23, 25, 30, 35, 40],
    'income': [35000, 40000, 50000, 60000, 70000]
})
```

```python
# Calcul du coefficient de corr√©lation de Pearson (lin√©aire)
pearson_corr, _ = pearsonr(data['age'], data['income'])
print("Coefficient de corr√©lation de Pearson :", pearson_corr)
```

```python
# Calcul du coefficient de corr√©lation de Spearman (monotone)
spearman_corr, _ = spearmanr(data['age'], data['income'])
print("Coefficient de corr√©lation de Spearman :", spearman_corr)
```
