## Les bases du Machine Learning

Le **Machine Learning (ML)** est une branche de l’intelligence artificielle qui permet aux ordinateurs d’apprendre à partir des données, sans être explicitement programmés.
Il est crucial pour les analystes de données, car il permet de créer des modèles prédictifs, détecter des patterns complexes, et automatiser la prise de décision basée sur les données.

---

## Applications dans divers secteurs

1. **Finance**
   * Détection de fraudes, scoring de crédit, trading algorithmique

2. **Santé**
   * Diagnostic assisté par IA, analyse d’images médicales, prédiction d’évolution de maladies

3. **Commerce / Marketing**
   * Recommandations de produits, segmentation de clientèle, prédiction de churn

---

## **Types de Machine Learning**

| Type                 | Définition                                                            | Exemple concret                                |
| -------------------- | --------------------------------------------------------------------- | ---------------------------------------------- |
| **Supervisé**        | Modèle entraîné avec des données labellisées (entrée + sortie connue) | Prédire si un client remboursera un prêt       |
| **Non supervisé**    | Modèle entraîné sans labels, découvre des structures cachées          | Regrouper des clients par comportement d’achat |
| **Par renforcement** | L’agent apprend par essais/erreurs en recevant des récompenses        | Robot qui apprend à sortir d’un labyrinthe     |

---

## **Processus typique de développement d’un modèle ML**

### 1. **Sélection des variables (Feature Selection)**

* Identifier les variables les plus pertinentes pour le problème.
Ex : Revenu, score de crédit, âge pour un modèle de prêt.

### 2. **Choix du modèle (Model Selection)**

* Sélection d’un algorithme adapté au type de problème : classification, régression, clustering, etc.
Ex : Régression logistique pour une classification binaire.

### 3. **Évaluation du modèle (Model Evaluation)**

* Mesurer la performance avec des métriques adaptées : accuracy, F1-score, AUC, etc.
* Utilisation de validation croisée pour garantir la robustesse.
